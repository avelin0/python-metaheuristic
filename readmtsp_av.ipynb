# READ FILE TO MATRIX DISTANCE
##  read file to array data structure

```python
import math
import pandas as pd

def readFile(filename):
    dataset = open(filename,"r")
    lines = dataset.readlines()
    lines[1].split()
    dataset.close()
    return lines[1::]

lines = readFile("class1/mtsp51_3.txt")
print(lines)
```

## read array to array of points

```python
# read array to array of points
def fromArrayToArrayOfTuples(lines):
    points = []
    for line in lines:
        mTuple = (line.split()[1],line.split()[2])
        points.append(mTuple)
    return points

points = fromArrayToArrayOfTuples(lines)
print(len(points))
print(points)
```

## function input points output matrix distance

```python
# function input points output matrix distance
# defMatrixDistance(points)
def fromPointsToDistanceMatrix(points):
    end = len(points) -1
    cols = rows = len(points)
    matrix = [[0 for _ in range(cols)] for _ in range(rows)]
    for i in range(0,rows):
        for j in range(0,cols):
            # print(i,j)
            xi = float(points[i][0])
            yi = float(points[i][1])
            xj = float(points[j][0])
            yj = float(points[j][1])
            distij = math.sqrt( (xi-xj)**2 + (yi-yj)**2 )
            matrix[i][j] = distij
    return matrix
distanceMatrix = fromPointsToDistanceMatrix(points)
print(distanceMatrix[0][1]) # test case - d01 = 12.36
print(distanceMatrix[49][50]) # test case - d4950 = 26.17
print(distanceMatrix[50][49]) # test case - d4950 = 26.17
print(distanceMatrix[48][49]) # test case - d4849 = 12.04
rows = len(distanceMatrix)
cols = len(distanceMatrix[0])
print(rows,cols)
```

```python
## read matrix Distance write in a File
```

```python
def frommatrixDistanceToFile(distanceMatrix, outputName):
    outputFile = open(outputName, "w+")
    for i in range(0,rows):
        for j in range(0,cols):
            outputFile.write(str(distanceMatrix[i][j]))
            if j == (rows - 1) :
                outputFile.write("\n")
            else:
                outputFile.write(" ")
    outputFile.close()
frommatrixDistanceToFile(distanceMatrix,"matrixDistance_output.txt")
```
